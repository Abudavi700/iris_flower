#Import Library

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

#Load data

```python
iris = load_iris(as_frame=True)
X = iris.data
y = iris.target

X.head()
```

```python
#lihat shape datanya
print("features shape: ", X.shape)
print("target shape: ", y.shape)
```

#Split dataset into train and test set

```python
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.7,
                                                    random_state=42)

#checking the shape
print("Number of training features: ", X_train.shape)
print("Number of testing features: ", X_test.shape)
print("Number of training target: ", y_train.shape)
print("Number of testing target: ", y_train.shape)
```

#Training model
this iris case is using KNN classifier model to train

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=3)
#train the model by feeding data to model
model.fit(X_train, y_train)
```

#Evaluating model

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print("The accuracy score of model: ", acc)
```

#Addition: utilizing the trained model

```python
new_data = [[5, 5, 3, 2],
            [2, 4, 3, 5]]

prediction = model.predict(new_data)
prediction
```

```python
prediction_species = [iris.target_names[p] for p in prediction]
print(prediction_species)
```
